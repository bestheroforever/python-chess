```python
import math 
  
# Function to calculate the Probability 
def Probability(rating1, rating2): 
  
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, 1.0 * (rating1 - rating2) / 400)) 
  
  
# Function to calculate Elo rating 
# K is a constant. 
# d determines whether 
# Player A wins or Player B.  
def EloRating(Ra, Rb, K, d): 
   
  
    # To calculate the Winning 
    # Probability of Player B 
    Pb = Probability(Ra, Rb) 
  
    # To calculate the Winning 
    # Probability of Player A 
    Pa = Probability(Rb, Ra) 
  
    # Case -1 When Player A wins 
    # Updating the Elo Ratings 
    if (d == 1) : 
        Ra = Ra + K * (1 - Pa) 
        Rb = Rb + K * (0 - Pb) 
      
  
    # Case -2 When Player B wins 
    # Updating the Elo Ratings 
    else : 
        Ra = Ra + K * (0 - Pa) 
        Rb = Rb + K * (1 - Pb) 
      
  
    return [int(Ra),int(Rb)]
  
# Driver code 
  
# Ra and Rb are current ELO ratings 

K = 40
```

```python
playernumber=9
players=[randomer,getfirstmove,alphabet,alphabetreverse,capturebot,checkbot,capturecheckbot,simpleevalutaion,simpleevalutaionwithcheck]
players_name=["randomer","getfirstmove","alphabet","alphabetreverse","capturebot","checkbot","capturecheckbot","simpleevalutaion","simpleevalutaionwithcheck"]
def Tournoment():
    
    for i in range(100):
        a=random.randint(0,playernumber-1)
        b=random.randint(0,playernumber-1)
        while a==b:
            b=random.randint(0,playernumber-1)
        Game(board,players[a],players[b])
        if board.result()=='1-0':
            score[a], score[b]=EloRating(score[a], score[b], K, 1) 
        elif board.result()=='0-1':
            score[a], score[b]=EloRating(score[a], score[b], K, 0) 
        if i%30==0:
            print(players_name[a],"  " ,board.result(),"  ",players_name[b])
            display(board)
    np.savetxt('AA_data.dat', score)
```
